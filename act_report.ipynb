WeRateDogs rates super-cute dog images. Dog owner send images of their little rascals (mostly dogs but not always) to be rated. The twitter feed is very amusing harmless fun. With 8.8 Mio followers WeRateDogs is also extremely successful. Looking at all the tweets in this feed, I asked myself, who is the cutest of them all? There is no right answer to this of course but looking at some of the tweets between 2015 and 2017 there are some interesting trends to be seen. 


## Who is the cutest of them all?

Many different dog types are rated on WeRateDogs. I wanted to know, which breed was rated the most and of the ones that were rated at least 20 times, which breed had the higest average rating. To find the answer, I used data that tries to predict the dog breed from the images that were attached to the tweets. 

**Result:**
From all the images with single dogs that were rated, the **Golden Retriever** was rated the most often. Not surprising - it's the familiy dog number 1! On average the most highest rated was the **Samoyed breed** and very handsome they are indeed.

The data that I looked at span 2 years. To see how consistent the ranking was I looked at a timeline of average monthly rankings. Did WeRateDogs rank the dogs consistently?

It looks like at the start of the twitter feed, the average ratings were around 10/10. With time either the dogs got cuter or the Raters more soppy. In July the average rating was up to 14/10 already. So don't feel bad, if your dog was rated lower early on in the twitter feed. There was a lot toughter rating then. In July 2016 and Apr 2017, I found some silly high rankings - a reason why I did not look for the highest ranked tweet - gimmick ratings as far as I can tell. 

Apart from two month (Jul 2016 and Apr 2017) with massive outliers, the mean rating of the dogs has increased with time. Starting around 10 and increasing to 13 in Sep 2017.
